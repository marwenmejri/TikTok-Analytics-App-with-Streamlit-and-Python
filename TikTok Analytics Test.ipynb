## Create Helper Function to Process Data

```python
import json
```

```python
# Load our initial json Data
with open("export.json", 'r') as f:
    data = json.load(f)
data
```

```python
nested_values = ['video', 'author', 'music', 'stats', 'authorStats']
skip_data = ["challenges", "duetInfo", "textExtra", "stickerOnItem"]
```

```python
flattened_data = {}
for idx, value in enumerate(data):
    new_dict = {}
    keys = value.keys()
    for prop_key, prop_value in value.items():
        if prop_key in nested_values:
            for proper_key_ , prop_value_ in prop_value.items():
                new_dict[f"{prop_key}_{proper_key_}"] = prop_value_
        elif key in skip_data:
            continue
        else:
            new_dict[prop_key] = prop_value
    flattened_data[idx] = new_dict
```
